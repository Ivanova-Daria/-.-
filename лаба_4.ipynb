# № 4.2.22

Даны n вещественных чисел. Определите максимальное и минимальное из них.

При выводе вещественных результатов оставьте два знака после запятой.

```
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_04_02_22.
#
# Выполнил: Иванова Д.Д.
# Группа: АДЭУ-201

n = int(input("n = "))
i = 1
a = []
while i != n+1:
  e = float(input("{}-е число ".format(i)))
  a.append(e)
  i += 1
a_max = max(a) 
a_min = min(a) 

print("Максимум: {:.2f}".format(a_max))
print("Минимум: {:.2f}".format(a_min))

# --------------
# Пример вывода:
#
# n = 4
# 1-е число = 6.2
# 2-е число = 3.8
# 3-е число = 1.1
# 4-е число = 9.66
# Максимум: 9.66
# Минимум: 1.10
```
